```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load CSV

df = pd.read_csv("../data/user_spending.csv")

df.head()
```

**Visualization 1 - Ages with  the highest income**
We are computing the average income by age and show top ages

# Average total income per age
age_income = df.groupby("age")["total_income"].mean().sort_values(ascending=False)

# View top ages
age_income.head()

Plotting as a bar chart

```python
plt.figure()
age_income.plot(kind="bar")
plt.xlabel("Age")
plt.ylabel("Average Total Income")
plt.title("Average Income by Age")
plt.tight_layout()
plt.savefig("../data/age_income_chart.png", dpi=300)
plt.show()
```

**Visualization 2 â€“ Gender distribution across spending categories**
We are calculating the average spending per category for each gender:

```python
expense_cols = [
    "utilities_amount",
    "entertainment_amount",
    "school_fees_amount",
    "shopping_amount",
    "healthcare_amount",
]

gender_expenses = df.groupby("gender")[expense_cols].mean()

gender_expenses
```

Plotting a grouped bar chart

```python
plt.figure()
gender_expenses.T.plot(kind="bar")
plt.xlabel("Expense Category")
plt.ylabel("Average Amount")
plt.title("Average Spending by Gender and Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../data/gender_spending_chart.png", dpi=300)
plt.show()
```
